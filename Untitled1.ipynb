```python
import pandas as pd

# Load the dataset
df = pd.read_csv('cleaned_bangalore_traffic_wards.csv')

# Ensure 'Date' column is converted to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Add time-related features
df['hour'] = df.index.hour
df['day_of_week'] = df.index.dayofweek
df['month'] = df.index.month

# Check the updated dataset with new features
print(df[['hour', 'day_of_week', 'month']].head())
```

```python
print(df.columns)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# Assuming df is already loaded and cleaned

# List of all features (including 'Bookings' as the target)
features = ['Traffic Volume', 'Average Speed', 'Travel Time Index', 'Congestion Level',
            'Road Capacity Utilization', 'Incident Reports', 'Environmental Impact',
            'Public Transport Usage', 'Traffic Signal Compliance', 'Parking Usage',
            'Pedestrian and Cyclist Count', 'Roadwork and Construction Activity',
            'latitude', 'longitude', 'h3_index', 'hour', 'day_of_week', 'month',
            'Weather Conditions_Fog', 'Weather Conditions_Overcast',
            'Weather Conditions_Rain', 'Weather Conditions_Windy', 'Bookings']  # Target

# Scale all the features using MinMaxScaler
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(df[features])

# Prepare input (X) and output (y) for training (now y contains all features)
X = []
y = []

timesteps = 30  # Number of timesteps (sequence length)
for i in range(len(scaled_features) - timesteps):
    X.append(scaled_features[i:i + timesteps])
    y.append(scaled_features[i + timesteps])  # Target is all features at next timestep

X, y = np.array(X), np.array(y)

# Split the dataset into training and testing sets
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build the LSTM-based model (can be swapped with TFT architecture)
model = Sequential()
model.add(LSTM(128, activation='relu', input_shape=(timesteps, len(features))))
model.add(Dense(len(features)))  # Output layer predicts all features

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f"Test MAE: {test_mae}")

# Make predictions for all features
predictions_scaled = model.predict(X_test)

# Inverse transform the predictions back to original scale (all features)
predictions_unscaled = scaler.inverse_transform(predictions_scaled)

# Inverse transform the actual y_test back to original scale for comparison
y_test_unscaled = scaler.inverse_transform(y_test)

# Show some predictions vs actual values (for all features)
print("Predictions vs Actual (for all features):")
for i in range(10):
    print(f"Predicted: {predictions_unscaled[i]}, Actual: {y_test_unscaled[i]}")

# Plot the comparison for one of the features (e.g., 'Bookings')
plt.plot(predictions_unscaled[:, features.index('Bookings')], label='Predicted Bookings')
plt.plot(y_test_unscaled[:, features.index('Bookings')], label='Actual Bookings')
plt.title('Predicted vs Actual Bookings')
plt.xlabel('Time')
plt.ylabel('Bookings')
plt.legend()
plt.show()
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler

# Define the model
model = Sequential()
model.add(LSTM(128, activation='relu', return_sequences=True, input_shape=(timesteps, len(features))))
model.add(LSTM(64, activation='relu', return_sequences=True))
model.add(LSTM(32, activation='relu'))
model.add(Dense(len(features)))  # Output layer

# Compile the model with Adam optimizer and learning rate
model.compile(optimizer=Adam(learning_rate=1e-4), loss='mse', metrics=['mae'])

# Early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Learning rate scheduler function
def scheduler(epoch, lr):
    if epoch % 10 == 0 and epoch != 0:
        return lr * 0.5  # Decrease the learning rate by half every 10 epochs
    return lr

lr_scheduler = LearningRateScheduler(scheduler)

# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2,
                    callbacks=[early_stopping, lr_scheduler])
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Define the model
model = Sequential()
model.add(LSTM(128, activation='relu', return_sequences=True, input_shape=(timesteps, len(features))))
model.add(LSTM(64, activation='relu', return_sequences=True))
model.add(LSTM(32, activation='relu'))
model.add(Dense(len(features)))  # Output layer

# Increase the learning rate slightly
initial_learning_rate = 1e-3  # Start with a higher learning rate

# Compile the model with Adam optimizer and the updated learning rate
model.compile(optimizer=Adam(learning_rate=initial_learning_rate), loss='mse', metrics=['mae'])

# Early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Use ReduceLROnPlateau instead of fixed scheduler
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)

# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2,
                    callbacks=[early_stopping, reduce_lr])
```

```python
import matplotlib.pyplot as plt

# Plot loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot MAE
plt.plot(history.history['mae'], label='Training MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.xlabel('Epochs')
plt.ylabel('Mean Absolute Error')
plt.legend()
plt.show()
```
